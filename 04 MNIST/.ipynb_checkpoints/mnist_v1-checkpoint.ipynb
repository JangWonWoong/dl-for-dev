```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

### MNIST 데이터셋 로딩
트레인셋 60,000개 테스트셋 10,000개로 분리
레이블은 자동으로 one-hot 엔코딩 처리됨

```
mnist = keras.datasets.mnist
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

X_train 데이터는 28x28 이미지 데이터가 60000개 있음 <br/>
데이터를 1차원 벡터로 변환 (60000, 784)

```
X_train = X_train.reshape(60000, 28*28)
```

```
X_train.shape
```

```
X_test = X_test.reshape(10000, 28*28)
X_test.shape
```

```
X_train = X_train.astype('float32')
X_train.dtype
```

```
X_test = X_test.astype('float32')
X_test.dtype
```

```
X_train
```

```
# [0, 1] 사이의 값으로 정규화 (Normalization)
X_train /= 256
X_test /= 256
```

```
# One-hot encoding
Y_train = tf.keras.utils.to_categorical(Y_train, 10)
Y_test = tf.keras.utils.to_categorical(Y_test, 10)
```

### 모델 만들기
입력층과 출력층 두 개의 층을 가진 네트워크

```
model = tf.keras.models.Sequential()
model.add(keras.layers.Dense(10, input_shape=(28*28,), activation='softmax'))
```

```
# 모델 요약보기
model.summary()
```

```
# 모델 컴파일
model.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
# 모델 학습 시키기
model.fit(X_train, Y_train, batch_size=128, epochs=200, 
          verbose=True, valication_split=0.2)
```

```
# 모델 평가
test_loss, test_acc = model.evaluate(X_test, Y_test)
print('정확도: ', test_acc, ' 오차: ', test_loss)
```
